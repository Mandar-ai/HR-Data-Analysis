```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Dataset Creation

data = {
    'Employee ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112, 113, 114, 115],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy', 'Karl', 'Lara', 'Mike', 'Nina', 'Oscar'],
    'Age': [28, 35, 30, 40, 25, 29, 34, 31, 45, 36, 27, 33, 50, 38, 41],
    'Department': ['IT', 'HR', 'Marketing', 'Finance', 'IT', 'HR', 'Marketing', 'Finance', 'IT', 'HR', 'Marketing', 'Finance', 'IT', 'HR', 'Marketing'],
    'Salary': [5500, 4800, 6200, 7000, 5300, 4600, 6400, 7100, 5600, 4700, 6300, 7200, 5800, 4900, 6500],
    'Joining Year': [2020, 2018, 2019, 2017, 2021, 2019, 2020, 2016, 2015, 2018, 2020, 2016, 2014, 2019, 2015],
    'Performance Rating': [4, 3, 5, 4, 3, 2, 5, 4, 3, 2, 5, 4, 3, 2, 5],
    'Leaves Taken': [8, 12, 5, 10, 6, 15, 4, 8, 9, 14, 3, 7, 10, 13, 2]
}

df = pd.DataFrame(data)

# Step 2: Perform Basic Analysis

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Show dataset information
print("\nDataset Information:")
print(df.info())

# Calculate the average salary and performance rating
avg_salary = df['Salary'].mean()
avg_rating = df['Performance Rating'].mean()
print(f"\nAverage Salary: ${avg_salary:.2f}")
print(f"Average Performance Rating: {avg_rating:.2f}")

# Count the number of employees in each department
dept_count = df['Department'].value_counts()
print("\nNumber of employees in each department:")
print(dept_count)

# Data Manipulation
# Add a new column 'Experience'
df['Experience'] = 2024 - df['Joining Year']

# Filter employees with a Performance Rating greater than 3
high_perf = df[df['Performance Rating'] > 3]
print("\nEmployees with Performance Rating > 3:")
print(high_perf)

# Find the highest-paid employee and their department
highest_paid = df.loc[df['Salary'].idxmax()]
print("\nHighest-paid employee details:")
print(highest_paid)

# Step 3: Visualization

# Bar Chart: Count of employees in each department
plt.figure(figsize=(8, 6))
df['Department'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()

# Histogram: Distribution of employee salaries
plt.figure(figsize=(8, 6))
df['Salary'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Box Plot: Salaries across different departments
plt.figure(figsize=(8, 6))
sns.boxplot(x='Department', y='Salary', data=df, palette='Set2')
plt.title('Comparison of Salaries Across Departments')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

# Scatter Plot: Relationship between Age and Salary
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Department', palette='viridis', s=100)
plt.title('Relationship Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

# Optional Challenge

# Department with the highest average performance rating
dept_high_perf = df.groupby('Department')['Performance Rating'].mean().idxmax()
print(f"\nDepartment with highest average performance rating: {dept_high_perf}")

# Analyze relationship between Performance Rating and Leaves Taken
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Performance Rating', y='Leaves Taken', data=df, hue='Department', palette='cool', s=100)
plt.title('Performance Rating vs Leaves Taken')
plt.xlabel('Performance Rating')
plt.ylabel('Leaves Taken')
plt.show()
```
